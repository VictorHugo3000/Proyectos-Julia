# Algoritmo de bisección
> Por Arturo Erdely

Considere una función continua $f$ sobre un intervalo cerrado $[a,b]$ tal que $f(a)f(b) < 0$ y suponga que existe un único valor $c$ en dicho intervalo tal que $f(c) = 0.$ Programaremos en `Julia` el *algoritmo de bisección* (ver referencias $[1]$ y $[2]$ al final) para aproximar numéricamente el valor de dicha raíz, especificando un error máximo de aproximación $\delta$ de dicha raíz (tolerancia) y un número máximo de iteraciones $m.$

Por continuidad existe un valor $z$ en el intervalo abierto $\,]a,b[\,$ tal que $f(z)=0.$ El algortimo consiste en dividir por la mitad iterativamente el intervalo inicial $[a,b]$ y en cada iteración seleccionar el subintervalo que contiene a la raíz $z.$ Esto es $z_{1}=(a+b)/2$ y si $f(z_{1})=0$ entonces $z = z_{1}$ y ya terminamos; y si no, entonces necesariamente alguno de los intervalos $[a,z_1]$ y $[z_1,b]$ contiene a $z$ ¿cuál? Pues aquél que cumpla la condición de que la función tenga signos opuestos al evaluarse en sus extremos, y así sucesivamente.

```julia
function bisección(f, a, b; δ = abs((a + b)/2) / 1_000_000, m = 10_000)
    iter = 1
    z = (a + b) / 2
    while iter ≤ m
        if f(z) == 0.0 || (b - a)/2 ≤ δ
            break
        elseif f(a) * f(z) > 0
            a = z
        else
            b = z
        end
        z = (a + b) / 2
        iter += 1
    end
    iter -= 1
    if iter == m
        println("Se alcanzó el número máximo de iteraciones = $m")
    end
    return (raíz = z, dif = f(z), numiter = iter, maxiter = m, tol = δ)
end
```

```julia
f(x) = (x - 3) * (x - 1) * (x + 1)
```

```julia
bisección(f, -2, -1.5)
```

¿Qué pasa si buscamos la raíz mediante *fuerza bruta*?

```julia
findmin([3, 4, 1, 7, 9]) #Es la contraparte de findmax
```

```julia
function fuerzaBruta(f, a, b; δ = abs((a + b)/2) / 1_000_000)
    n = Int(round(ceil((b - a) / δ)))
    z = collect(range(a, b, length = n))
    v = abs.(f.(z))
    r = z[findmin(v)[2]]
    return (raíz = r, dif = f(r), tol = δ)
end
```

```julia
fuerzaBruta(f, -1.9, 0)
```

```julia
bisección(f, -1.9, 0)
```

```julia
function fuerzaBruta2(f, a, b; δ = abs((a + b)/2) / 1_000_000)
    n = Int(round(ceil((b - a) / δ)))
    Δ = (b - a) / n
    r = a 
    v = abs(f(r))
    for i ∈ 1:n
        r2 = a + i*Δ
        v2 = abs(f(r2))
        if v2 < v
            r = r2
            v = v2
        end
    end
    return (raíz = r, dif = f(r), tol = Δ)
end
```

```julia
fuerzaBruta2(f, -1.9, 0)
```

```julia
@btime fuerzaBruta2(f, -1.9, 0)
```
