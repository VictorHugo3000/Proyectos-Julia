```julia
fechaInicio = Date(2021,5,1)
fechaFinal = fecha - Day(14) # omitir últimas 2 semanas (por reazgo en información)
serieFechas = collect(fechaInicio:Day(1):fechaFinal)
```

```julia
# series
sPos = zeros(Int, length(serieFechas))
sNeg = zeros(Int, length(serieFechas))
sMue = zeros(Int, length(serieFechas))

# porcentajes

pPos = zeros(length(serieFechas)) # tasa de positividad
pMue = zeros(length(serieFechas)) # tasa de letaldad
```

```julia
i = 0
for día ∈ serieFechas
    i += 1
    iFecha = findall(df.FECHA_SINTOMAS .== día)
    
    iFechaPos = iFecha ∩ iPos; sPos[i] = length(iFechaPos)
    iFechaNeg = iFecha ∩ iNeg; sNeg[i] = length(iFechaNeg)
    iFechaMue = iFecha ∩ iMue; sMue[i] = length(iFechaMue)

    pPos[i] = 100 * sPos[i] / (sPos[i] + sNeg[i])  #Tasa de positividad
    pMue[i] = 100 * sMue[i] / sPos[i] #Tasa de letalidad
end
```

```julia
m = [1,2,3,20,5,6,7,8,9,10]

sum(m[1:4])
```

```julia
promóvil = function(v, d)
    n = length(v)
    pm = zeros(n - d + 1)
    for i ∈ d:n
        pm[i - d + 1] = sum(v[(i - d + 1):i]) / d
    end
    return pm
end
```

```julia
d = 7
P = promóvil(sPos, d) ./ 1000
N = promóvil(sNeg, d) ./ 1000
plot(serieFechas[d:end], P, lw = 3, color = :red, 
    label = "positivos", yticks = 0:5:50, legend = (0.3, 0.9))
plot!(serieFechas[d:end], N, lw = 3, color = :green, label = "negativos")
title!("Casos positivos y negativos")
xaxis!("@ArturoErdely || COVID19mx hasta el " * string(fechaFinal))
yaxis!("Miles de personas, prom. móvil $d días")
```

```julia
d = 7
M = promóvil(sMue, d)
plot(serieFechas[d:end], M, lw = 3, color = :gray, 
    label = "positivos", yticks = 0:10:250, legend = false)
title!("Defunciones confirmadas")
xaxis!("@ArturoErdely || COVID19mx hasta el " * string(fechaFinal))
yaxis!("Núm. personas, prom. móvil $d días")
```

```julia
d = 7
tpos = promóvil(pPos, d) 
tlet = promóvil(pMue, d) 
plot(serieFechas[d:end], tpos, lw = 3, color = :red, 
     label = "positividad", yticks = 0:5:100, legend = (0.2, 0.8))
title!("Positividad y letalidad")
xaxis!("@ArturoErdely || COVID19mx hasta el " * string(fechaFinal))
yaxis!("Porcentaje en prom. móvil $d días")
hline!([0], lw = 0.05, color = :gray, label ="")
plot!(serieFechas[d:end], tlet, lw = 3, color = :purple, label = "letalidad",)
```
